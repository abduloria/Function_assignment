```python

# 8-3 T-Shirt
def make_shirt(size, message):
    print(f"Shirt size: {size}, Message: {message}")

# Call using positional arguments
make_shirt("Large", "I love Python")
# Call using keyword arguments
make_shirt(size="Medium", message="Hello, World!")

# 8-4 Large Shirts
def make_shirt(size="Large", message="I love Python"):
    print(f"Shirt size: {size}, Message: {message}")

make_shirt()
make_shirt("Medium")
make_shirt("Small", "Different message")

# 8-5 Cities
def describe_city(city, country="Default Country"):
    print(f"{city} is in {country}")

describe_city("Reykjavik", "Iceland")
describe_city("Paris")
describe_city("New York", "USA")

# 8-6 City Names
def city_country(city, country):
    return f"{city}, {country}"

print(city_country("Santiago", "Chile"))
print(city_country("Tokyo", "Japan"))
print(city_country("Berlin", "Germany"))

# 8-7 Album
def make_album(artist, title, songs=None):
    album = {'artist': artist, 'title': title}
    if songs:
        album['songs'] = songs
    return album

album1 = make_album("Artist1", "Album1", 12)
album2 = make_album("Artist2", "Album2")
album3 = make_album("Artist3", "Album3", 8)

print(album1)
print(album2)
print(album3)

# 8-8 User Albums
while True:
    artist = input("Enter artist (or 'quit' to exit): ")
    if artist.lower() == 'quit':
        break
    title = input("Enter album title: ")
    album = make_album(artist, title)
    print(album)

# 8-9 Messages
def show_messages(messages):
    for message in messages:
        print(message)

messages_list = ["Message 1", "Message 2", "Message 3"]
show_messages(messages_list)

# 8-10 Sending Messages
def send_messages(messages, sent_messages):
    for message in messages:
        print(message)
        sent_messages.append(message)

sent_messages_list = []
send_messages(messages_list, sent_messages_list)
print(messages_list)
print(sent_messages_list)

# 8-11 Archived Messages
archived_messages = messages_list.copy()
send_messages(archived_messages, sent_messages_list)
print(messages_list)
print(sent_messages_list)

# 8-12 Sandwiches
def make_sandwich(*ingredients):
    print("Sandwich with:")
    for ingredient in ingredients:
        print(f"- {ingredient}")

make_sandwich("Ham", "Cheese")
make_sandwich("Turkey", "Lettuce", "Tomato")

# 8-13 User Profile
def build_profile(first_name, last_name, **additional_info):
    profile = {'first_name': first_name, 'last_name': last_name}
    profile.update(additional_info)
    return profile

my_profile = build_profile("John", "Doe", age=25, location="City", occupation="Developer")
print(my_profile)

# 8-14 Cars
def make_car(manufacturer, model, **options):
    car_info = {'manufacturer': manufacturer, 'model': model}
    car_info.update(options)
    return car_info

car = make_car('subaru', 'outback', color='blue', tow_package=True)
print(car)

# 8-15 Printing Models
# Save functions in a separate file named printing_functions.py
# Import and use them in printing_models.py

# 8-16 Imports
import module_name
from module_name import function_name
from module_name import function_name as fn
import module_name as mn
from module_name import *

# 8-17 Styling Functions
# Ensure functions follow the styling guidelines described in the chapter.
```
