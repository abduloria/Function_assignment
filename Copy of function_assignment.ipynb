```python

# 8-7. Album
def make_album(artist, title, songs=None):
    album = {"artist": artist, "title": title}
    if songs:
        album["songs"] = songs
    return album

# Make three dictionaries representing different albums
print(make_album("Artist1", "Album1", 10))
print(make_album("Artist2", "Album2"))
print(make_album("Artist3", "Album3", 15))
```

```python

# 8-9. Messages
def show_messages(messages):
    for message in messages:
        print(message)

# Make a list of short text messages
messages = ["Why are you here? ", "Create to empower", "Discipline is better than motivation "]

# Call the function to show messages
show_messages(messages)
```

```python

# 8-10. Sending Messages
def send_messages(messages):
    sent_messages = []
    for message in messages:
        print(message)
        sent_messages.append(message)
    return sent_messages

# Call the function and print both lists
sent_messages = send_messages(messages)
print("Original Messages:", messages)
print("Sent Messages:", sent_messages)
```

```python

# 8-11. Archived Messages
archived_messages = send_messages(messages.copy())
print("Original Messages:", messages)
print("Archived Messages:", archived_messages)
```

```python

# 8-12. Sandwiches
def make_sandwich(*ingredients):
    print(f"Sandwich with: {', '.join(ingredients)}")

# Call the function with different numbers of arguments
make_sandwich("Cheese", "Tomato")
make_sandwich("Ham", "Lettuce", "Mayo")
make_sandwich("Turkey", "Swiss", "Mustard", "Pickles")
```

```python

# 8-13. User Profile
def build_profile(first_name, last_name, **additional_info):
    profile = {"first_name": first_name, "last_name": last_name}
    profile.update(additional_info)
    return profile

# Build a profile of yourself
my_profile = build_profile("Musa", "Ahmed", age=25, occupation="Engineer", location="Abuja")

print(my_profile)
```

```python

# 8-14. Cars
def make_car(manufacturer, model, **kwargs):
    car_info = {"manufacturer": manufacturer, "model": model}
    car_info.update(kwargs)
    return car_info

# Call the function with required and optional information
car = make_car('subaru', 'outback', color='blue', tow_package=True)
print(car)
```
